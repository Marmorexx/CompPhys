```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#default plot settings function
def plot_settings (title,xlabel,ylabel):
    plt.figure(figsize=(8,8))
    plt.title(title, fontsize=14)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.grid(True)
    return;
```

```python
#return the coefficients for the population equation as an array to fit in the solver
def getcoeffs(rho, kappa):
    return np.array([rho/kappa, -rho, 1+rho/kappa, -rho])

#replace zeros with nan's. Used to plot properly
def zero_to_nan(values):
    return [float('nan') if x==0 else x for x in values]

#same algorithm for the three root solutions
def zero_to_nan_2d(values):
    for i in range(0, len(values)):
        for j in range(0, len(values[i])):
            if (values[i][j]==0):
                values[i][j]=float('nan')
    return values

#check if solutions are stable or unstable by derivative
def stability_check(n, rho, kappa):
    check = rho-2*rho/kappa*n - 2*n/(n**2 + 1)**2
    if (check < 0):
        print('{:.3f}'.format(n), "This solution is stable")
    elif (check > 0):
        print('{:.3f}'.format(n), "This solution is unstable")
```

```python
#define variables we'll need for calculations
g_kappa = 7.5
rho_space = np.zeros(20)
rho = 0
step = 0.05

#fill rho array with evenly spaced numbers
for i in range(0, len(rho_space)):
    rho_space[i] = rho
    rho = rho + step
    
#initialise two arrays for the one root and three root solutions
real_roots = np.zeros(len(rho_space))
real_triple_roots = np.zeros((len(rho_space), 3))

#loop over all values for rho and determine the roots for each resulting polynomial
for i in range(0, len(rho_space)):
    pol = getcoeffs(rho_space[i], g_kappa)
    roots = np.roots(pol)
    
    #filter for real values by neglecting all imaginary parts smaller than 1e-5 
    #to account for numerical uncertainties that might make the imaginary part nonzero
    real_valued = roots.real[abs(roots.imag)<1e-5]
    
    #fill arrays for one root and three root solutions respectively
    if(len(real_valued)==1):
        real_roots[i] = real_valued
    elif(len(real_valued)==3):
        for j in range(0, 3):
            real_triple_roots[i][j] = real_valued[j]

#plot variables
plot_settings("Stationary solutions of the population equation", r"$\rho$", "$n^*$")

#convert zeros to nan's
real_roots_plot = zero_to_nan(real_roots)
real_triple_roots_plot = zero_to_nan_2d(real_triple_roots)

#Triple roots are different solutions for the same rho, as can be seen in the plot
plt.plot(rho_space, real_roots_plot, label="Single roots")
plt.plot(rho_space, real_triple_roots_plot, label="Triple roots")
plt.legend(loc='best')
plt.savefig("ex62.pdf")
```

```python
#check all solutions for stability respectively
for i in range(0, len(rho_space)):
    stability_check(real_roots_plot[i], rho_space[i], g_kappa)
for i in range(0, len(rho_space)):
    for j in range(0, 3):
        stability_check(real_triple_roots_plot[i][j], rho_space[i], g_kappa)
```
